```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
month = 6
data_raw_dir = f"E:\\Data\\BigDataLecture\\raw\\2024-0{month}.csv"
data_pro_dir = f"E:\\Data\\BigDataLecture\\processed\\2024-0{month}_processed.csv"
data_all_dir = f"E:\\Data\\BigDataLecture\\processed\\data.csv"
```

### 读取数据

```python
data_raw = pd.read_csv(data_raw_dir)
data_raw_value = data_raw.values

data_pro = pd.read_csv(data_pro_dir)
data_pro_value = data_pro.values

data_all = pd.read_csv(data_all_dir)
data_all_value = data_all.values
```

### 画图

```python
plt.figure(figsize=(12, 6))
for i in range(8):
    plt.subplot(2, 4, i + 1)
    plt.title(f'Feature {i + 1}: {data_pro.columns[i]}')
    plt.xlabel('Time')
    plt.ylabel('Value')

    # Plot processed data
    plt.plot(data_pro_value[:, i], label='Processed Data', color='orange')

    # Plot raw data
    if i == 0:
        plt.legend()

plt.tight_layout()
```

```python
plt.figure(figsize=(12, 6))
for i in range(8):
    plt.subplot(2, 4, i + 1)
    plt.title(f'Feature {i + 1}: {data_all.columns[i]}')
    plt.xlabel('Time')
    plt.ylabel('Value')

    # Plot processed data
    plt.plot(data_all_value[:, i], label='Processed Data', color='orange')

    # Plot raw data
    if i == 0:
        plt.legend()

plt.tight_layout()
```

```python
import os
months = [1, 2, 3]
data_folder = "E:\\Data\\BigDataLecture\\processed"
data_list = [os.path.join(data_folder, f"2024-0{month}_processed.csv") for month in months]
data = None
for data_path in data_list:
    tmp = pd.read_csv(os.path.join(data_folder, data_path))
    if data is not None:
        data = pd.concat((data, tmp), ignore_index=True)
    else:
        data = tmp
```
